<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">4.2 <span style="color:#F50057;">تعریف توابع</span></h1> <a id='3'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">تاکنون از توابع از پیش تعریف شده زیادی استفاده کرده‌اید. توابعی نظیر 
    (int, float, print, input, type, sum, len, min و max)
    و توابع 
    (mean, median و mode)
    در آمار توصیفی، توابعی از این دست محسوب می‌شوند. هر یک از آنها یک وظیفه واحد و کاملاً مشخصی را انجام می‌دادند. شما اغلب توابع سفارشی را تعریف و فراخوانی می‌کنید. کدهای ذیل یک تابع به نام 
    square
    را تعریف می‌کند، این تابع مربع آرگومان ورودی را محاسبه می‌کند. این تابع دوبار فراخوانی می‌شود، در بار اول با آرگومان عدد صحیح ۷ فراخوانی می‌شود و عدد صحیح ۴۹ را تولید می‌کند، در بار دوم با عدد اعشاری ۲.۵ فراخوانی می‌شود و عدد اعشاری ۶.۲۵ را تولید می‌کند:
</div>

````python
def square(number):
    """Calculate the square of number."""
    return number ** 2
````

````python
square(7)
````

````python
square(2.5)
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">بازگرداندن یک نتیجه در فراخوانی یک تابع</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">وقتی یک تابع اجرا را تمام می‌کند، کنترل را به فراخواننده خود برمی‌گرداند، فراخواننده در حقیقت دستوری است که تابع را فراخوانی می‌کند. در بلوک 
    square
    ، دستور 
    return
    :
</div>

```Python
return number ** 2
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">ابتدا مربع

    را محاسبه می‌کند، سپس تابع را خاتمه می‌دهد و در نهایت نتیجه را به فراخواننده برمی‌گرداند. در این مثال اولین فراخواننده در اسنیپت [۲] است، بنابراین 
    IPython
    نتیجه را در 
     Out[2]
    نشان می‌دهد. فراخواننده دوم در اسنیپت [۳] است، بنابراین 
    IPython

    نمایش می‌دهد.
</div>

    None
    فقدان یک مقدار را نشان می‌دهد. مقدار 
    None

    ارزیابی می‌شود.<br>

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">4.3 <span style="color:#F50057;">توابع با چندین پارامتر</span></h1> <a id='4'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در این قسمت یک تابع به نام 
    maximum
    تعریف می‌کنیم. این تابع بزرگترین مقدار را از میان سه مقدار تعیین کرده و آن را باز می‌گرداند، در ادامه این تابع را با اعداد صحیح، اعداد اعشاری و رشته‌ها سه بار فراخوانی می‌کنیم.
</div>

````python
def maximum(value1, value2, value3):
    """Return the maximum of three values."""
    max_value = value1
    if value2 > max_value:
        max_value = value2
    if value3 > max_value:
        max_value = value3
    return max_value
````

````python
maximum(12, 27, 36)
````

````python
maximum(12.3, 45.6, 9.7)
````

````python
maximum('yellow', 'red', 'orange')
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">پرتاب یک تاس ۶۰۰۰۰۰۰ بار</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">اگر 
    محدوده
    واقعاً اعداد صحیح را به صورت تصادفی تولید کند، هر عدد در محدوده خود هر بار که آن را فراخوانی می‌کنیم، احتمال (یا 
    chance
    یا 
    likelihood
    ) برابری دارد که برگردانده شود. 
    برای نشان دادن این‌که اعداد روی وجوه تاس احتمال‌های برابر دارند، اسکریپت بعدی پرتاب تاس را شبیه‌سازی می‌کند. مطابق نمونه خروجی، هنگامی‌که اسکریپت را اجرا می‌کنید، تقریبا هر عدد روی تاس ۱۰۰۰۰۰۰ بار می‌آید.
</div>

````python
### Rolling a Six-Sided Die 6,000,000 Times

# fig04_01.py
"""Roll a six-sided die 6,000,000 times."""
import random

# face frequency counters
frequency1 = 0
frequency2 = 0
frequency3 = 0
frequency4 = 0
frequency5 = 0
frequency6 = 0

# 6,000,000 die rolls
for roll in range(6_000_000):  # note underscore separators
    face = random.randrange(1, 7)

    # increment appropriate face counter
    if face == 1:
        frequency1 += 1
    elif face == 2:
        frequency2 += 1
    elif face == 3:
        frequency3 += 1
    elif face == 4:
        frequency4 += 1
    elif face == 5:
        frequency5 += 1
    elif face == 6:
        frequency6 += 1

print(f'Face{"Frequency":>13}')
print(f'{1:>4}{frequency1:>13}')
print(f'{2:>4}{frequency2:>13}')
print(f'{3:>4}{frequency3:>13}')
print(f'{4:>4}{frequency4:>13}')
print(f'{5:>4}{frequency5:>13}')
print(f'{6:>4}{frequency6:>13}')
````

````python
# fig04_02.py
"""Simulating the dice game Craps."""
import random

def roll_dice():
    """Roll two dice and return their face values as a tuple."""
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    return (die1, die2)  # pack die face values into a tuple

def display_dice(dice):
    """Display one roll of the two dice."""
    die1, die2 = dice  # unpack the tuple into variables die1 and die2
    print(f'Player rolled {die1} + {die2} = {sum(dice)}')

die_values = roll_dice()  # first roll
display_dice(die_values)

# determine game status and point, based on first roll
sum_of_dice = sum(die_values)

if sum_of_dice in (7, 11):  # win
    game_status = 'WON'
elif sum_of_dice in (2, 3, 12):  # lose
    game_status = 'LOST'
else:  # remember point
    game_status = 'CONTINUE'
    my_point = sum_of_dice
    print('Point is', my_point)

# continue rolling until player wins or loses
while game_status == 'CONTINUE':
    die_values = roll_dice()
    display_dice(die_values)
    sum_of_dice = sum(die_values)

    if sum_of_dice == my_point:  # win by making point
        game_status = 'WON'
    elif sum_of_dice == 7:  # lose by rolling 7
        game_status = 'LOST'

# display "wins" or "loses" message
if game_status == 'WON':
    print('Player wins')
else:
    print('Player loses')
````

````python
def rectangle_area(length, width):
    """Return a rectangle's area."""
    return length * width
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">تعریف یک تابع با لیست اختیاری از آرگومان‌ها</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">بیایید یک تابع به نام 
    average
    تعریف کنیم که می‌تواند تعداد دلخواهی از آرگومان‌ها را قبول کند:
</div>

````python
def average(*args):
    return sum(args) / len(args)
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">نام پارامتر 
    args
    بر حسب قرارداد استفاده شده است ولی شما می‌توانید از هر شناسه دیگری استفاده کنید. اگر تابع دارای چندین پارامتر باشد، پارامتر 
    args*
    باید راست‌ترین پارامتر باشد. اکنون تابع 
    average
    را چندین بار و هر بار با لیست‌هایی با طول‌های مختلف فراخوانی می‌کنیم:
</div>

````python
average(5, 10)
````

````python
average(5, 10, 15)
````

````python
average(5, 10, 15, 20)
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">برای محاسبه میانگین، مجموع عناصر 
    args tuple
    (که توسط تابع درونی 
    sum
    برگردانده می‌شود) را بر تعداد عناصر 
    tuple
    تقسیم می‌کنیم (که توسط تابع درونی 
    len
    برگردانده می‌شود). اگر تعداد 
    args
    در تعریف تابع 
    average
    صفر باشد، خطای 
    ZeroDivisionError 
    رخ می‌دهد. در فصل بعد، نحوه دسترسی به یک 
    tuple
    را بدون بازکردن آن خواهید دید.
</div>

    tuple
    ، یک لیست و هر داده نوع

    اعمال شود، تمامی عناصر آن را از هم جدا می‌کند. کد زیر یک لیست پنج عنصری به نام

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">4.15 <span style="color:#F50057;">ارسال آرگومان‌ها به توابع: نگاهی عمیق‌تر</span></h1> <a id='16'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">بیایید نگاهی دقیق‌تر به نحوه ارسال آرگومان‌ها به بیندازیم. در بسیاری از زبان‌های برنامه نویسی، دو روش برای ارسال آرگومان وجود دارد، ارسال با مقدار و ارسال با ارجاع (که گاهی اوقات فراخوانی با مقدار و فراخوانی با ارجاع نامیده می‌شود):

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">

</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">آرگومان‌های پایتون همیشه با ارجاع منتقل می‌شوند. از آنجایی که همه چیز در پایتون شیء محسوب می‌شود، بسیاری از افراد به این روش، ارسال شیء با ارجاع می‌گویند. اگر هنگام فراخوانی تابع یک آرگومان تولید شود، پایتون یک ارجاع به شیء آرگومان و نه خود شیء را در پارامتر مربوطه کپی می‌کند. این رویکرد برای کارایی بسیار مهم است. غالبا توابع با اشیاء بزرگی کار می‌کنند و کپی کردن مکرر این اشیاء باعث مصرف زیاد حافظه کامپیوتر و کاهش کارایی برنامه به طور چشمگیری می‌شود. حتی توابعی که در این فصل تعریف کرده‌اید و یا کلاس‌هایی(انواع سفارشی) که در فصل‌های بعدی تعریف خواهید کرد، یک شیء در پایتون محسوب می‌شوند.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">تابع از پیش تعریف شده id و هویت اشیاء</h2> <a id=''></a>

    ، برای به‌دست آوردن یک مقدار صحیح منحصر به فرد برای شناسایی یک شیء استفاده کنیم. تابع

    فقط برای اشیایی که در حافظه قرار دارند به کار می‌رود (وقتی این را روی کامپیوتر خود اجرا می‌کنید احتمالا نتایج متفاوتی خواهید گرفت):

    می‌توانید ثابت کنید که در پایتون آرگومان و پارامتر هر دو به یک شیء اشاره می‌کنند، اگر دو عملوند عملگر 
    is
    دارای شناسه‌ای برابر باشند، مقدار 
    True
    را برمی‌گرداند:
</div>

````python
def cube(number):
    print('number is x:', number is x)  # x is a global variable
    return number ** 3
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">اشیاء غیر قابل تغییر به عنوان آرگومان
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">هنگامی که یک تابع آرگومانی را به عنوان ارجاع به یک شیء تغییر ناپذیر (یا غیر قابل تغییر) مانند 
    int
    ، float، 
    string
    یا 
    tuple
    دریافت می‌کند، حتی اگر دسترسی مستقیم به شیء اصلی فراخواننده را هم داشته باشید، نمی‌توانید مقدار شیء غیرقابل تغییر اصلی را تغییر دهید. برای اثبات این موضوع، از طریق یک دستور انتساب اضافی مقدار شناسه 
    number
    در تابع 
    cube 
    را قبل و بعد از اختصاص به یک شیء جدید نشان می‌دهیم:
</div>

````python
def cube(number):
    print('id(number) before modifying number:', id(number))
    number **= 3
    print('id(number) after modifying number:', id(number))
    return number
````

````python
cube(x)
````

    cube(x)

    print

    id(number) 
    با شناسه 
    id(x)
    در اسنیپت [۲] برابر است. مقادیر عددی غیر قابل تغییر محسوب می‌شوند، بنابراین دستور ذیل:

    number
    اختصاص می‌دهد. به یاد بیاورید که اگر هیچ ارجاعی به یک شیء اصلی وجود نداشته باشد، از حافظه جمع آوری خواهد شد

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">4.16 <span style="color:#F50057;">Recursion</span></h1> <a id='17'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در این قسمت قصد داریم یک برنامه برای محاسبه یک تابع مشهور ریاضی به نام فاکتوریل بنویسیم. فاکتوریل عدد صحیح مثبت 
    n
    که به صورت 
    $ n! $
    نوشته شده و فاکتوریل 
    n
    خوانده می‌شود، حاصل ضرب اعداد طبیعی است از ۱ تا خود 
    n:
    <br>
</div>

$ n · (n – 1) · (n – 2) · … · 1 $

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    $ 1! $
    برابر ۱ بوده و بنا به تعریف 
    $ 0! $
    برابر ۱ است. برای مثال 
    $ 5! $
    حاصل ضرب 
    $ 5 · 4 · 3 · 2 · 1 $
    بوده و برابر ۱۲۰ است.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">رویکرد تکراری فاکتوریل

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">حل مسائل به روش بازگشتی

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">رویکردهای بازگشتی حل مسئله‌ها دارای چندین مولفه مشترک هستند. هنگامی که یک تابع بازگشتی را برای حل یک مسئله فراخوانی می‌کنید، قادر است مسئله را تنها در ساده‌ترین حالت و یا حالت پایه حل کند. اگر تابع را با یک حالت پایه فراخوانی کنید، بلافاصله نتیجه را باز می‌گرداند. اگر تابع را با یک مسئله پیچیده‌تر فراخوانی کنید، معمولا مسئله را به دو قسمت تقسیم می‌کند: قسمت اول که تابع چگونگی انجام آن را می‌داند و قسمت دوم که نحوه انجام آن را نمی‌داند. برای عملی بودن روش بازگشتی، قسمت دوم باید کمی از مسئله اصلی ساده‌تر یا کوچک‌تر باشد. از آنجا که این مسئله جدید شبیه مسئله اصلی است، تابع یک کپی جدید از خود را برای کار بر روی مسئله کوچک‌تر فراخوانی می‌کند. این فراخوانی اخیر به عنوان فراخوانی بازگشتی و یا گام بازگشت نامیده می‌شود. مفهوم تفکیک مسئله به دو بخش کوچک‌تر شکلی از رویکرد تقسیم و راهبرد است که قبلا مطرح شده است.
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">مرحله بازگشتی تا زمانی که فراخوانی تابع اصلی فعال باشد اجرا می‌شود (یعنی اجرا کامل نشده است). از آنجایی که تابع هر زیر مسئله جدید را به دو بخش مفهومی تقسیم می‌کند ممکن است فراخوانی بازگشتی زیادی انجام شود. برای پایان یافتن بازگشت، باید تابع هر بار خود را با یک نسخه ساده‌تر از مسئله اصلی فراخوانی کند و توالی این مسئله‌های کوچک‌تر و کوچک‌تر باید در نهایت به سمت حالت پایه همگرا شود. هنگامی که تابع حالت پایه را تشخیص دهد، نتیجه را به نسخه قبلی تابع برمی‌گرداند. دنباله‌ای از بازگشت‌ها پیش می‌آید و این فرآیند آن‌قدر تکرار می‌شود تا در نهایت فراخوانی اصلی تابع، مقدار نهایی را به فراخواننده بازگرداند.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">رویکرد بازگشتی فاکتوریل
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">با توجه به آنکه رابطه 
    $ n! $
    را می‌توان مطابق فرمول ذیل نوشت، به راحتی می‌توانید به نمایش بازگشتی فاکتوریل برسید:
</div>    
    <br>
    $ n! = n · (n – 1)! $
    <br>
<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">    
    برای مثال 
    $ 5! $
    برابر 
    $ 5 · 4! $
    می‌باشد:
</div>    
    <br>
    $ 5! = 5 · 4 · 3 · 2 · 1 $
    <br>
    $ 5! = 5 · (4 · 3 · 2 · 1) $
    <br>
    $ 5! = 5 · (4!) $


<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">دیاگرام بازگشتی
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">محاسبه 
    $ 5! $
    مانند شکل زیر انجام می‌گیرد. ستون سمت چپ نشان می‌دهد که چگونه توالی فراخوانی بازگشتی تا 
    $ 1! $
    ادامه می‌یابد (حالت پایه). در این مرحله مقدار ۱ ارزیابی شده و بازگشت خاتمه می‌یابد. ستون سمت راست مقادیری که در فراخوانی بازگشتی به فراخواننده بازمی‌گرداند را از پایین به بالا نشان می‌دهد. این فرآیند تا زمانی که مقدار نهایی محاسبه و برگردانده شود، ادامه خواهد داشت.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">پیاده‌سازی بازگشتی تابع فاکتوریل
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">قطعه کد زیر برای محاسبه و نمایش فاکتوریل اعداد صحیح ۰ تا ۱۰ از روش بازگشتی استفاده می‌کند:
</div>

````python
def factorial(number):
    """Return factorial of number."""
    if number <= 1:
        return 1
    return number * factorial(number - 1) # recursive call
````

````python
for i in range(11):
    print(f'{i}! = {factorial(i)}')
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">تابع بازگشتی فاکتوریل در اسنیپت [۴] ابتدا تعیین می‌کند که آیا شرط خاتمه یعنی نامساوی 
    number <= 1
    درست است و یا خیر. اگر این شرط درست باشد (حالت پایه)، 
    factorial
    عدد ۱ را برمی‌گرداند و دیگر نیازی به بازگشت بیشتر نیست. اگر 
    number
    بزرگتر از ۱ باشد دستور 
    return
    دوم مسئله را به صورت حاصل ضرب 
    number
    و یک فراخوانی بازگشتی 
    factorial
    بیان می‌کند. فراخوانی اخیر مقدار 
    factorial(number - 1)
    را محاسبه می‌کند. این فراخوانی در قیاس با محاسبه 
    factorial(number)
    مسئله کوچک‌تری محسوب می‌شود. توجه داشته باشید که فاکتوریل تابع باید مقادیر غیر منفی دریافت کند. این الزام در تعریف تابع فاکتوریل بررسی نمی‌شود.
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">حلقه تکرار در اسنیپت [۵] تابع 
    factorial
    را برای مقادیر ۰ تا ۱۰ فراخوانی می‌کند. خروجی نشان می‌دهد که مقادیر فاکتوریل به سرعت رشد می‌کنند. پایتون بر خلاف بسیاری از زبان‌های برنامه نویسی برای اندازه یک عدد صحیح محدودیتی قائل نمی‌شود.
</div>

</div>

<br>
<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">تابع 
     range
    و دستور 
    for
    جزئیات حلقه تکرار کنترل شده با شمارنده را پنهان می‌کنند. شما فقط دامنه مقادیر مورد نظر و متغیری که باید مقادیر تولید شده را دریافت کند، تعیین می‌کنید. تابع

    می‌داند که چگونه این مقادیر را تولید کند. به همین ترتیب دستور 
    for

</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">توابع داخلی پایتون مانند 
    sum, min و max
    هر کدام از تکرارهای داخلی استفاده می‌کنند. برای جمع لیست نمرات، به سادگی آنچه را که می‌خواهید انجام دهید، اعلان می‌کنید؛ یعنی

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">4.18 <span style="color:#F50057;">مقدمه‌ای بر علم داده: سنجش پراکندگی</span></h1> <a id='19'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">برای تعیین واریانس با میانگین این مقادیر یعنی ۳.۵ شروع می‌کنیم. این نتیجه با تقسیم مجموع مقادیر یعنی ۳۵، بر تعداد آنها یعنی ۱۰ به دست می‌آید. در مرحله بعد، میانگین را از هر مقدار کم می‌کنیم (که برخی نتایج منفی تولید می‌کنند):
</div>
$ -2.5,\ -0.5,\ 0.5,\ -1.5,\ 2.5,\ 1.5,\ -0.5,\ 0.5,\ 1.5,\ -1.5 $

<br>
<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در ادامه، هر یک از این نتایج را به توان ۲ می‌رسانیم (نتیجه حاصل شده فقط اعداد مثبت هستند).
</div>

$ 6.25,\ 0.25,\ 0.25,\ 2.25,\ 6.25,\ 2.25,\ 0.25,\ 0.25,\ 2.25,\ 2.25 $

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">برای سادگی، واریانس جمعیت را محاسبه می‌کنیم. تفاوت ظریفی بین واریانس جمعیت و واریانس نمونه وجود دارد. واریانس نمونه به جای تقسیم کردن بر 
    n
    (تعداد پرتاب تاس در مثال ما) بر 
    n - 1
    تقسیم می‌شود. این تفاوت برای نمونه‌های کوچک مشهود است و با افزایش تعداد نمونه‌ها ‌بی‌اهمیت می‌شود. دو تابع 
    pvariance
    و 
    variance
    در ماژول 
    statistics
    به ترتیب برای محاسبه واریانس جمعیت و واریانس نمونه به کار می‌روند. به طور مشابه، در ماژول 
    statistics
    توابع 
    pstdev
    و 
    stdev
    برای محاسبه انحراف معیار جمعیت و انحراف معیار نمونه استفاده می‌شوند.
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در نهایت برای محاسبه واریانس جمعیت، میانگین اعداد بالا را 
   ( $ {22.5} / {10} = {2.25} $ )
    محاسبه می‌کنیم. مربع کردن تفاضلات اختلاف هر مقدار با میانگین، بر داده‌های پرت تاکید می‌کند. داده‌های پرت مقادیری هستند که از میانگین فاصله زیادی دارند. هرچه بیشتر به تحلیل داده‌ها می‌پردازیم، گاهی اوقات می‌خواهیم به داده‌های پرت دقیق توجه کنیم و گاهی اوقات آن‌ها را نادیده می‌گیریم. کد زیر برای تایید محاسبه دستی ما از تابع 
    pvariance
    در ماژول 
    statistics
    استفاده می‌کند:
</div>

````python
import statistics
````

````python
statistics.pvariance([1, 3, 4, 2, 6, 5, 3, 4, 5, 2])
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">انحراف معیار
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">انحراف استاندارد جذر واریانس است (در این حالت ۱.۵) و تاثیر داده‌های پرت را کاهش می‌دهد. هرچه واریانس و انحراف استاندارد کوچک‌تر باشد، داده‌ها به میانگین نزدیک‌ترند و پراکندگی کلی (یعنی گسترش) بین مقادیر و میانگین کمتر است. کد زیر انحراف معیار جمعیت را با تابع 
    pstdev
    در ماژول 
    statistics
    محاسبه می‌کند و صحت نتیجه دستی ما را تایید می‌کند:
</div>

````python
statistics.pstdev([1, 3, 4, 2, 6, 5, 3, 4, 5, 2])
````
