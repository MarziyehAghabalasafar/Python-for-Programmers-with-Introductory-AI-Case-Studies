</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در فصل قبل، تولید اعداد تصادفی و شبیه‌سازی پرتاب یک تاس شش وجهی را نشان دادیم. این فصل را با مقدمه‌ای بر علم داده به پایان می‌رسانیم، از کتابخانه‌های

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">استفاده از عملگر 
    =+
    برای الصاق به لیست</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">اجازه دهید با یک لیست خالی 
     []
    شروع کنیم، سپس با استفاده از دستور 
    for
    و عملگر 
    =+
    مقادیر ۱ تا ۵ را به لیست اضافه کنیم، لیست برای جا دادن هر داده به صورت پویا رشد می‌کند:
</div>

````python
a_list = []
````

````python
for number in range(1, 6):
    a_list += [number]
````

````python
a_list
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">وقتی عملوند سمت چپ عملگر 
    =+
    یک لیست باشد، عملوند سمت راست باید از نوع 
    iterable
    باشد؛ در غیر این صورت، یک خطای 
    TypeError
    رخ می‌دهد. در کدهای اسنیپت [۱۹]، کروشه‌های دور متغیر 
    number
    یک لیست تک‌عضوی ایجاد می‌کند و ما آن را به 
    a_list
    اضافه می‌کنیم. اگر عملوند سمت راست عملگر 
    =+
    شامل چندین عضو باشد، این عملگر همه آن‌ها را اضافه می‌کند. در کدهای زیر کاراکترهای 
    'Python'
    به لیست 
    letters
    اضافه می‌شوند:
</div>

````python
letters = []
````

````python
letters += 'Python'
````

````python
letters
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">اگر عملوند سمت راست 
    =+
    یک تاپل باشد، عناصر آن نیز به لیست اضافه می‌شوند. در ادامه و در همین فصل، از متد 
    append
    برای اضافه کردن اقلام دیگر به لیست استفاده خواهیم کرد.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">الحاق لیست‌ها با عملگر +
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">با استفاده از عملگر 
    +
    می‌توانید دو لیست، دو تاپل یا دو رشته را به هم پیوند دهید. نتیجه، یک دنباله جدید از همان نوع است که اجزای آن از عناصر عملوند سمت چپ و به دنبال آن عناصر عملوند سمت راست تشکیل می‌شوند. دنباله‌های اصلی تغییر نخواهند کرد:
</div>

````python
list1 = [10, 20, 30]
````

````python
list2 = [40, 50]
````

````python
concatenated_list = list1 + list2
````

````python
concatenated_list
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">اگر عملوندهای عملگر 
    +
    از یک نوع نباشند، یک خطای 
    TypeError
    اتفاق می‌افتد. به عنوان مثال، اتصال یک لیست و یک تاپل خطا محسوب می‌شود.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">استفاده از 
    for
    و 
    range
    برای دسترسی به اندیس لیست و مقادیر آن
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">عناصر لیست همچنین از طریق اندیس‌ها و عملگر 
    subscription
    ([]) قابل دسترسی هستند:
</div>

````python
for i in range(len(concatenated_list)):  
    print(f'{i}: {concatenated_list[i]}')
````

````python
# fig05_01.py
"""Displaying a bar chart"""
numbers = [19, 3, 15, 7, 11]

print('\nCreating a bar chart from numbers:')
print(f'Index{"Value":>8}   Bar')

for index, value in enumerate(numbers):
    print(f'{index:>5}{value:>8}   {"*" * value}')
````

    ایجاد می‌کند و تعداد کاراکترها برابر 
    value
    است. هنگامی که عملگر 
    *
    با یک دنباله مورد استفاده قرار گیرد، دنباله یاد شده به اندازه 
    value
    تکرار می‌شود. در مثال بالا، دنباله فوق رشته

ارسال لیست‌ها به توابع
</span></h1> <a id='8'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در فصل قبلی به این نکته اشاره کردیم که همه اشیاء به وسیله ارجاع ارسال می‌شوند و ارسال یک شیء قابل تغییر را به عنوان آرگومان تابع نشان دادیم. در اینجا در مورد ارجاع بیشتر بحث می‌کنیم. برای این منظور بررسی می‌کنیم هنگام ارسال یک لیست تغییر پذیر به یک تابع چه رویدادهایی رخ می‌دهد.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">
ارسال کل یک لیست به تابع
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">تابع 
    modify_elements
    را در نظر بگیرید؛ این تابع یک ارجاع به یک لیست را دریافت می‌کند و هر یک از مقادیر اعضای لیست را در ۲ ضرب می‌کند:
</div>

````python
def modify_elements(items):
    """"Multiplies all element values in items by 2."""
    for i in range(len(items)):
        items[i] *= 2
````

````python
numbers = [10, 3, 7, 1, 9]
````

````python
modify_elements(numbers)
````

````python
numbers
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">پارامتر 
    items
    در تابع 
    modify_elements
    یک ارجاع به لیست اصلی را دریافت می‌کند و دستورات بدنه حلقه، هر عضو لیست اصلی را تغییر می‌دهند.
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">هنگامی که یک تاپل را به یک تابع ارسال می‌کنید، هرگونه تلاش برای تغییر عناصر غیرقابل تغییر تاپل، منجر به خطای 
    TypeError
    می‌شود:

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    traceback
    قبلی، دو اسنیپتی که منجر به 
    TypeError
    شده‌اند را نشان می‌دهد. اولین 
    traceback
    مربوط به فراخوانی تابع در اسنیپت [۷] و دومین 
    traceback
    مربوط به تعریف تابع در اسنیپت [۱] است. شماره خط قبل از کد اسنیپت قرار می‌گیرد. تاکنون بیشتر اسنیپت‌هایی که نشان داده‌ایم، تک‌خطی بودند. وقتی یک استثنا در چنین اسنیپتی روی دهد، همیشه قبل از آن 
      1 <---- 
    قرار می‌گیرد و نشان می‌دهد که خط ۱ (تنها خط اسنیپت) باعث استثنا شده است. اسنیپت‌های چندخطی مانند تعریف تابع 
    modify_elements
    شماره خط‌های متوالی که با ۱ شروع می‌شوند را نشان می‌دهند. 
      4 <----
    در کد بالا نشان می‌دهد که استثنا در خط ۴ 
    modify_elements
    رخ داده است. مهم نیست که 
    traceback
    چقدر طولانی است، همیشه استثنا در آخرین خطی که نماد 
       <----
    قرار دارد روی می‌دهد.

<h1 style="direction:rtl;font-size: 35px;font-family: 'XB Zar';color:#8e24aa;">5.13 <span style="color:#F50057;">
Generator Expressions
</span></h1> <a id='14'></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    generator expression
    شبیه 
    list comprehension
    است، اما یک شیء 
    generator
    نوع 
    iterable
    ایجاد می‌کند و این شیء به نوبه خود مقادیر مورد نیاز را تولید می‌کند. این روش 
    lazy evaluation
    نامیده می‌شود. روش 
    list comprehension
    از رویکرد حریصانه 
    (greedy evaluation)
    استفاده می‌کند و لیست‌ها را بلافاصله هنگام اجرای آن‌ها ایجاد می‌کند. هنگامی که تعداد داده‌ها زیاد باشد، ایجاد یک لیست می‌تواند حافظه و زمان قابل توجهی بگیرد؛ بنابراین اگر در یک زمان کل لیست مورد نیاز نباشد، استفاده از 
    generator expression
    می‌تواند میزان مصرف حافظه برنامه شما را کاهش دهد و کارایی آن را بهبود بخشد.
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    Generator expression
    همان توانایی 
    list comprehension
    را دارد، اما باید آنها را به جای براکت در پرانتز تعریف کنید. 
    generator expression
    در اسنیپت [۲] مربع مقادیر فرد در 
    numbers
    را برمی‌گرداند:
</div>

````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
````

````python
for value in (x ** 2 for x in numbers if x % 2 != 0):
    print(value, end='  ')
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">برای نشان دادن اینکه 
    generator expression
    یک لیست ایجاد نمی‌کند، بیایید 
    generator expression
    اسنیپت قبلی را به یک متغیر اختصاص دهیم و متغیر را ارزیابی کنیم:
</div>

````python
squares_of_odds = (x ** 2 for x in numbers if x % 2 != 0)
````

````python
squares_of_odds 
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">
فیلتر کردن مقادیر دنباله با تابع داخلی 
    filter
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در این قسمت برای به دست آوردن مقادیر فرد در 
    numbers
    از تابع داخلی 
    filter
    استفاده می‌کنیم:
</div>

````python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
````

````python
def is_odd(x):
    """Returns True only if x is odd."""
    return x % 2 != 0
````

````python
list(filter(is_odd, numbers))
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در پایتون توابع مانند داده‌ها شیء محسوب می‌شوند و می‌توانید آنها را به متغیرها اختصاص دهید، به توابع دیگر ارسال کنید و آنها را از توابع برگردانید. توابعی که توابع دیگر را به عنوان آرگومان دریافت می‌کنند، یکی از قابلیت‌های برنامه نویسی به سبک تابعی هستند و توابع مرتبه بالاتر 
     (higher-order functions)
    نامیده می‌شوند. به عنوان مثال، اولین آرگومان 
    filter
    باید تابعی باشد که یک آرگومان دریافت کند و اگر مقدار باید در نتیجه لحاظ شود، 
    True
    را برمی‌گرداند. اگر آرگومان ورودی تابع 
    is_odd
    عددی فرد باشد، مقدار 
    True
    را برمی‌گرداند. تابع 
    filter
    برای هر مقدار در آرگومان دوم خود 
    (numbers)
    که از نوع 
    iterable
    است، تابع 
    is_odd
    را فراخوانی می‌کند. ممکن است توابع مرتبه بالاتر یک تابع را به عنوان نتیجه برگردانند.
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">تابع 
    filter
    یک 
    iterator
    برمی‌گرداند، بنابراین تا زمانی که نتایج 
    filter
    را پیمایش نکنید، این نتایج تولید نمی‌شوند. این حالت نمونه دیگری از 
    lazy evaluation
    است. در اسنیپت [۳] تابع 
    list
    نتایج را پیمایش کرده و لیستی حاوی این نتایج را ایجاد می‌کند. با روش 
    list comprehension
    و استفاده از دستور 
    if
    به همین نتایج خواهیم رسید:
</div>

````python
[item for item in numbers if is_odd(item)]
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">
استفاده از لامدا به جای تابع
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">برای توابع ساده‌ای مانند 
     is_odd
    که فقط مقدار یک عبارت ساده را برمی‌گرداند، می‌توانید از یک عبارت لامدا استفاده کنید. با این رویکرد می‌توانید تابع را در جایی که لازم است به صورت 
    inline
    تعریف کرده و به یک تابع دیگر ارسال کنید:
</div>

````python
list(filter(lambda x: x % 2 != 0, numbers))
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">مقدار بازگشتی تابع 
    filter
    (یک iterator)
    را به تابع 
    list
    ارسال می‌کنیم، این تابع نتایج را به یک لیست تبدیل کرده و آنها را نشان می‌دهد. یک عبارت لامدا در حقیقت یک تابع بدون نام
    (anonymous function)
    محسوب می‌شود. در فراخوانی تابع 
    filter
    :
</div>

```Python
filter(lambda x: x % 2 != 0, numbers)
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">اولین آرگومان، یک عبارت لامدا است:
</div>

```Python
lambda x: x % 2 != 0
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">یک عبارت لامدا با کلمه کلیدی 
    lambda
    شروع می‌شود به دنبال آن لیست پارامترها که با کاما جدا شده‌اند قرار می‌گیرد، در ادامه کاراکتر دو نقطه 
    (:)
    و در انتها عبارت محاسباتی قرار می‌گیرد. در این مثال لیست پارامترها فقط یک پارامتر به نام 
     x
    دارد. یک عبارت لامدا به طور ضمنی مقدار محاسباتی خود را برمی‌گرداند و به همین دلیل بسیاری از توابعی که به شکل زیر تعریف شده‌اند
</div>


```Python
def function_name(parameter_list):     return expression
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">را می‌توان با عبارت لامدا ساده‌تر تعریف کرد:
</div>


```Python
lambda parameter_list: expression
```

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">
نگاشت مقادیر یک دنباله به مقادیر جدید
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در این قسمت با استفاده از تابع 
    map
    و یک عبارت لامدا، مقادیر دنباله 
    numbers
    را به توان ۲ می‌رسانیم:
</div>

````python
numbers
````

````python
list(map(lambda x: x ** 2, numbers))
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">اولین آرگومان تابع 
    map
    یک مقدار را دریافت می‌کند و مقدار جدیدی برمی‌گرداند، در این مثال، این آرگومان یک عبارت لامدا است که آرگومان خود را مربع می‌کند. آرگومان دوم، داده‌های از نوع 
    iterable
    برای تابع 
    map
    است. تابع 
    map
    از 
    lazy evaluation
    استفاده می‌کند. مقداری که تابع 
    map
    برمی‌گرداند را به تابع 
     list
    ارسال می‌کنیم، این مقدار از نوع 
    iterator
    است. این رویکرد به ما این امکان را می‌دهد که لیستی از مقادیر نگاشت شده ایجاد کنیم و آن‌ها را پیمایش کنیم. کد زیر معادل ایجاد یک لیست به روش 
    list comprehension
    است:
</div>

````python
[item ** 2 for item in numbers]
````

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">
تلفیق 
    filter
    و 
    map
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">مطابق کد زیر می‌توان توابع 
    filter
    و 
    map
    را با یکدیگر تلفیق کرد:
</div>

````python
list(map(lambda x: x ** 2, 
         filter(lambda x: x % 2 != 0, numbers)))
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در اسنیپت [۹] کدهای زیادی وجود دارد، بنابراین نگاهی دقیق به آن می‌اندازیم. در ابتدا 
    filter
    یک 
    iterable
    را برمی‌گرداند، این 
    iterable
    فقط مقادیر فرد در 
    numbers
    را نشان می‌دهد. تابع 
     map
    یک 
    iterable
    که مربع مقادیر فیلتر شده است را برمی‌گرداند. در نهایت تابع 
    list
    از 
    iterable
     تابع 
     map
    یک لیست ایجاد می‌کند. ممکن است کد زیر را به کد قبلی ترجیح دهید:
</div>

````python
[x ** 2 for x in numbers if x % 2 != 0]
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">به ازای هر مقدار 
    x
    در 
    numbers
    ، دستور 
     x ** 2
    فقط در صورت 
    True
    بودن شرط زیر ایجاد می‌شود:
</div>

```Python
x % 2 != 0
```

    min
    و

    آرگومان کلیدی تابع را برای هر عضو فراخوانی می‌کنند و از نتایج آن برای مقایسه عناصر استفاده می‌کنند.

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">
ترکیب 
    iterable
    ها در تاپل‌ها
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">تابع درونی 
    zip
    این امکان را به شما می‌دهد در یک زمان چندین داده 
    iterable
    را پیمایش کنید. این تابع هر تعداد 
    iterable
    را به عنوان آرگومان دریافت می‌کند و یک 
    iterator
    را برمی‌گرداند. این 
    iterator
    تعدادی تاپل با همان اندیس عناصر تولید می‌کند. برای مثال اسنیپت [۱۱] با فراخوانی تابع 
    zip
    تاپل‌های 
    ('Bob', 3.5)
    ، ('Sue', 4.0)
    و 
    ('Amanda', 3.75)
    را ایجاد کرده و به ترتیب آن‌ها را در اندیس‌های 0، 1 و 2 یک لیست قرار می‌دهد:
</div>

````python
names = ['Bob', 'Sue', 'Amanda']
````

````python
grade_point_averages = [3.5, 4.0, 3.75] 
````

````python
for name, gpa in zip(names, grade_point_averages):
    print(f'Name={name}; GPA={gpa}')
````

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">هر یک از تاپل‌ها را به 
    name
    و 
    gpa
    تفکیک کرده و نمایش می‌دهیم. کوتاه‌ترین آرگومان تابع 
    zip
    تعداد تاپل تولید شده را تعیین می‌کند. در اینجا طول هر دو یکسان است.
</div>

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">

    را مقداردهی می‌کنند.<br>

    a[2][3], a[2][2], a[2][1],  a[2][0]
    را مقداردهی می‌کنند.<br>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">یک لیست با

    ردیف و

    ستون یک لیست

    در

    نامیده می‌شود و

    <br>
    دستور 
    for
    تودرتوی زیر سطرهای لیست دوبعدی بالا را نشان می‌دهد. دستور

    اول هر بار یک سطر را نشان می‌دهد:

<h2 style="direction:rtl;font-size: px;font-family: 'XB Zar';color:#0288d1;">
نحوه اجرای حلقه‌های تودرتو
</h2> <a id=''></a>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">در این قسمت حلقه‌ تودرتو را به گونه‌ای تغییر می‌دهیم تا نام لیست، اندیس‌های سطر و ستون و مقدار هر یک از اعضا را نشان دهد:
</div>

````python
for i, row in enumerate(a):
    for j, item in enumerate(row):
        print(f'a[{i}][{j}]={item} ', end=' ')
    print()
````
